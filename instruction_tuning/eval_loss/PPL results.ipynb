```python
import os, sys
import pandas as pd
import glob
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfs = []
for path in glob.glob('*.csv'):
    df = pd.read_csv(path)
    if df['model'].values[0] == 'falcon-7b':
        for model in ['falcon-7b-lang-1', 'falcon-7b-lang-2', 'falcon-7b-lang-3','falcon-7b-lang-5', 'falcon-7b-lang-10', 'falcon-7b-lang-20', 'falcon-7b-lang-45']:
            df = df.copy()
            df['model'] = model
            dfs.append(df)
    else:
        dfs.append(df)
```

```python
df = pd.concat(dfs)
```

```python
sns.lineplot(data=df[df['checkpoint'] <= 20000], x='checkpoint', y='loss', hue='model')
plt.legend(loc='upper right')
```
