```python
import pandas as pd
import glob
```

```python
dfs = []
for path in glob.glob('*falcon*.csv'):
    model = path.split('eng_')[-1][:-4]
    if model[-1] == 'b':
        model_name = model
        lang = 0
    else:
        model_name = '-'.join(model.split('-')[:-1])[:-5]
        lang = int(model.split('-')[-1])
    df = pd.read_csv(path)
    df['model'] = model_name
    df['lang'] = lang
    dfs.append(df)
df = pd.concat(dfs)
```

```python
metric = 'macro_f1_score'
df.groupby(['model', 'lang', 'dataset'])[metric].mean().reset_index().pivot(index=['model', 'lang'], columns='dataset', values=metric)
```
